### CamelCase y snake_case

Python combina los estilos `snake_case` y `CamelCase` en su sintaxis:

*   __snake_case:__ letras_minusculas_separadas_por_underscore `my_variable`, `my_function`, `my_module`
*   __CamelCase:__ Primera letra de cada palabra Mayuscúla `ClassName`, `MethodName`, `MyVariable`.

* __Constantes__: TODO_EN_MAYUSCULA `PI = 3.1415`, `ALL_UPPERCASE = True`


Para la declaración de variables, funciones, clases, etc. es una buena práctica seguir las recomendaciones [PEP8](https://realpython.com/python-pep8/#naming-styles). En general, usaremos CamelCase para los nombres de las clases y snake_case para todo lo demás

###   __Operadores de asignación__

Podemos realizar los siguientes operadores para realizar asignaciones a una variable, es decir, operar sobre la misma variable


```python
x = 5
print(f"x = {x} \t x = 5")
x += 3
print(f"x += 3 \t x = x + 3 = {x}")
x -= 3
print(f"x -= 3 \t x = x - 3 = {x}")
x *= 3
print(f"x *= 3 \t x = x * 3 = {x}")
x /= 3
print(f"x /= 3 \t x = x / 3 = {x}")
x %= 3
print(f"x %= 3 \t x = x % 3 = {x}")
x //= 3
print(f"x //= 3  x = x // 3 = {x}")
x **= 3
print(f"x **= 3  x = x ** 3 = {x}")
```

### __Expresiones booleanas__

Podemos definir una expresión lógica como una proposición que tendrá un valor de verdad: cierto o falso. Para construir proposiciones podemos utilizar los operadores $\le, ≥, =, \lt, \gt, \neq$.

```python
a =  5
print(a < 3) #Menor que
print(a <= 3) #Menor o igual que
print(a > 3) #Mayor que
print(a >= 3) #Mayor o igual que
print(a == 3) #Igual que
print(a != 3) #Diferente de
```

También podemos generar condiciones más complejas utilizando las operadores lógicas `and`,` or` y `not`

```python
a = 0 > 3
b = 5 >= 2
c = 2 == 2.0
# operaciones logicas: and, or, not
print(a)
print(b)
print(c)
print(a and b and c)
print(a or b or c)
print(not(a))
print(not(b))
print(not(a and b and c))
```

Todos los tipos de datos tiene un equivalente en booleano. Este idea está relacionada con los valores que python considera `Falsy`


```python
falsy_values = (0, False, None, '', [], {})
print('0 -> boolean:', bool(0))
print('None -> boolean:', bool(None))
print("'' -> boolean:", bool(''))
print('[] -> boolean:', bool([]))
print('{} -> boolean:', bool([]))
```

## Condicionales

Los condicionales permiten tomar decisiones en el flujo de un programa. Dependiendo de si una condición es verdadera o falsa, el programa ejecutará un bloque de código específico.

<h1><center>Condicionales</center></h1>

Los condicionales nos permiten controlar el flujo de ejecución de un programa a partir del cumplimiento de una condición. Al evaluar dicha condición el resultado debe ser siempre `Verdadero` o `Falso`.

if: Evalúa si una condición es verdadera.

elif: Evalúa condiciones adicionales si la anterior no se cumplió.

else: Ejecuta un bloque si ninguna de las condiciones anteriores se cumple.

### __Indentación__

En Python, la indentación es de caracter obligatorio y se refiere al número de espacios al principio de una línea de código. Esta nos permite separar los bloques de código y mantener una mejor estructura y legibilidad  de nuestros programas (usualmente se usan 2 espacios para indicar un nuevo bloque de instrucciones)

### Tipos de condicionales

Python permite crear condicionales de tres tipos principales. Cada uno tiene una forma particular de evaluar las condiciones y tomar decisiones en el flujo del programa.

 1. If Statement (Condicional Simple)

Evalúa una condición. Si es True, ejecuta el bloque de código correspondiente. Si es False, no hace nada.

```python
# if condicion:
    # Código a ejecutar si la condición es verdadera
```

```python
edad = 18

if edad >= 18:
    print("Eres mayor de edad.")
```

2. If-Else Statement (Condicional con Alternativa)

Evalúa una condición. Si es True, ejecuta el bloque del if; si es False, ejecuta el bloque del else.

```python
 if condicion:
    # Código si la condición es verdadera
 else:
    # Código si la condición es falsa
```

```python
edad = 18

if edad >= 18:
    print("Eres mayor de edad.")
else:
    print("Eres menor de edad.")
```

 3. Nested If (Condicional Anidado)

Permite anidar condicionales dentro de otros condicionales. Es útil cuando necesitas evaluar múltiples condiciones dentro de una estructura.

```python
if condicion1:
    if condicion2:
        # Código si ambas condiciones son verdaderas
    else:
        # Código si la condicion2 es falsa
else:
    # Código si la condicion1 es falsa
```

```python
edad = 18
ciudad = "Bogota"

if edad >= 18:
    if ciudad == "Bogotá":
        print("Eres mayor de edad y vives en Bogotá.")
    else:
        print("Eres mayor de edad pero no vives en Bogotá.")
else:
    print("Eres menor de edad.")
```

4.if-elif-else

permite evaluar múltiples condiciones en cadena, puedes agregar tantas condiciones como necesites utilizando elif.
Si ninguna de las condiciones es verdadera, el bloque else se ejecuta.

```python
if condicion_1:
    # Código si la condicion_1 es verdadera
elif condicion_2:
    # Código si la condicion_2 es verdadera
elif condicion_3:
    # Código si la condicion_3 es verdadera
else:
    # Código si ninguna condición es verdadera
```

Su funcionamiento es el siguiente:

Se evalúa la primera condición (if).

Si es True, ejecuta el bloque de código correspondiente y termina la evaluación.

Si la primera condición es False, evalúa la siguiente (elif).

Si ninguna condición es True, se ejecuta el bloque else.

```python
calificacion = 95

if calificacion >= 90:
    print("Excelente")
elif calificacion >= 70:
    print("Aprobado")
elif calificacion >= 50:
    print("Necesitas mejorar")
else:
    print("Reprobado")
```

En python se puede utilizar condicionales con contracciones que dejan mas compactos los condicionales, a estos se les llama condicional ternario

```python
condicion = False

print("valor_si_verdadero" if condicion else "valor_si_falso")

print("valor_si_verdadero") if condicion else print("valor_si_falso")
```

```python
a = -2
print("Es positivo") if a > 0 else print("No es positivo")
```
