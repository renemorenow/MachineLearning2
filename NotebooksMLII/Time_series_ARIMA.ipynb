```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima.model import ARIMA

```

```python
y=[4]
np.random.seed(123)
for i in range(1, 100):
    y.append(0.4 * y[i-1] + 2 + np.random.normal(0, 3))

y = np.array(y)
```

```python
plt.plot(y)
```

```python
model = ARIMA(y, order=(1, 0, 0))
```

```python
model_fit = model.fit()
```

```python
model_fit.summary()
```

```python
y_predict = model_fit.predict()
```

```python
plt.plot(y, label = 'Real obs')
plt.plot(y_predict, label = 'Pred obs')
plt.legend()
plt.show()
```

```python
plot_pacf(y)
```

```python
plot_acf(y)
```

## MA de 2 rezagos

```python
np.random.seed(7)

errores = np.random.normal(0,1,300)
media = 30
y=[]
for t in range(1, 300):
    y.append(media +0.3*errores[t-1]+0.1*errores[t])

y = np.array(y)
```

```python
plt.plot(y)
```

```python
plot_pacf(y)
```

```python
plot_acf(y)
```

```python
model = ARIMA(y, order=(1, 0, 1))
```

```python
model_fit = model.fit()
```

```python
model_fit.summary()
```

```python
plt.plot(y)
plt.plot(model_fit.predict())
```

```python
plt.plot(model_fit.forecast(20))
```
